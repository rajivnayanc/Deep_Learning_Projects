```
import pandas as pd

data=pd.read_csv('iris data with class label.csv', header=None, names=list('ABCDE'))
data.head()
```

```
data_att=data.drop('E',axis=1)
data_label=data['E']
```

```
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(data_att, data_label)
data_label_pred = knn.predict(data_att)
print(metrics.accuracy_score(data_label, data_label_pred))
```

```
data_pca=pd.read_csv('iris data without class label.csv', names=list('ABCD'))
data_pca.head()
```

```
from sklearn.decomposition import PCA

pca = PCA(n_components=1)

data_pca_1=pca.fit_transform(data_pca)
```

```
data_pca_1=pd.DataFrame(data_pca_1, columns=list('A'))
```

```
# data_pca_1['classes']=data_label
# data_pca_1.head()
```

```
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(data_pca_1, data_label)
data_label_pred = knn.predict(data_pca_1)
print(metrics.accuracy_score(data_label, data_label_pred))
```
